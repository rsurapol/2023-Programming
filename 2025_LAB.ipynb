1. **การหาค่าความชัน (Slope) ของกราฟเส้น**
   - หากเรามีข้อมูลที่เป็นกราฟเส้นและต้องการหาค่าความชันของกราฟ (เช่น ความเร็วจากการเปลี่ยนแปลงของตำแหน่ง) เราสามารถใช้ `numpy.polyfit()` เพื่อคำนวณค่าความชันจากข้อมูล:
     ```python
     # ข้อมูลการเปลี่ยนแปลงของตำแหน่งตามเวลา
     time = np.array([0, 1, 2, 3, 4, 5])
     position = np.array([0, 2, 4, 6, 8, 10])

     # หาค่าความชันของกราฟ
     slope, intercept = np.polyfit(time, position, 1)  # 1 หมายถึงกราฟเส้นตรง
     ```

#### **ส่วนที่ 2: การสร้างสัญญาณพื้นฐาน**
1. สร้างสัญญาณไซน์ (Sine Wave) โดยใช้ `NumPy`:
   ```python
   # สร้างสัญญาณไซน์
   fs = 1000  # ความถี่ในการสุ่มตัวอย่าง (Hz)
   t = np.linspace(0, 1, fs, endpoint=False)  # เวลา 1 วินาที

   sine_wave = np.sin(2 * np.pi * f * t)

   plt.ylabel('Amplitude')
   plt.grid(True)
   plt.show()
   ```

#### **ส่วนที่ 3: การแปลง Fourier (Fourier Transform)**
1. **การแปลงสัญญาณเป็นความถี่**:
   ใช้การแปลง Fourier เพื่อดูว่าเรามีความถี่อะไรบ้างในสัญญาณ
   ```python
   # แปลง Fourier Transform
   fft_signal = np.fft.fft(sine_wave)
   freqs = np.fft.fftfreq(len(t), 1/fs)

   plt.plot(freqs[:fs//2], np.abs(fft_signal)[:fs//2])
   plt.title('Fourier Transform of Sine Wave')
   plt.xlabel('Frequency [Hz]')
   plt.ylabel('Magnitude')
   plt.grid(True)
   plt.show()
   ```

   - คำอธิบาย: จากกราฟนี้เราจะเห็นว่ามีการแสดงความถี่ที่ 50 Hz ซึ่งตรงกับความถี่ของสัญญาณไซน์ที่เราสร้างขึ้น

#### **ส่วนที่ 4: การกรองสัญญาณ (Signal Filtering)**
1. **การใช้ฟิลเตอร์ Low-pass**:
   ฟิลเตอร์ Low-pass ใช้เพื่อลดความถี่สูงในสัญญาณ
   ```python
   # สร้างฟิลเตอร์ Low-pass
   nyquist = 0.5 * fs  # Nyquist frequency
   low_cutoff = 100  # ความถี่ตัดที่ 100 Hz
   normal_cutoff = low_cutoff / nyquist
   b, a = signal.butter(4, normal_cutoff, btype='low', analog=False)

   # ใช้ฟิลเตอร์ Low-pass กับสัญญาณ
   filtered_signal = signal.filtfilt(b, a, sine_wave)

   # แสดงผล
   plt.plot(t, sine_wave, label='Original Signal')
   plt.plot(t, filtered_signal, label='Filtered Signal', linewidth=2)
   plt.title('Low-pass Filtered Signal')
   plt.xlabel('Time [s]')
   plt.ylabel('Amplitude')
   plt.legend()
   plt.grid(True)
   plt.show()
   ```

   - คำอธิบาย: จะเห็นว่าหลังจากกรองด้วยฟิลเตอร์ Low-pass สัญญาณที่มีความถี่สูงเกินกว่าค่าที่กำหนดจะถูกลดลงไป

#### **ส่วนที่ 2: การจำลองการเคลื่อนที่ของวัตถุในระบบฟิสิกส์**
สมมุติว่าเรามีวัตถุที่เคลื่อนที่ในแนวราบโดยไม่ถูกแรงต้าน และเราต้องการจำลองการเคลื่อนที่นี้

1. **สมการการเคลื่อนที่**:  
   เราจะใช้สมการการเคลื่อนที่ในแนวราบ ซึ่งสามารถเขียนได้ว่า:
$x(t) = x_0 + v_0 t + \frac{1}{2} a t^2$

โดยที่:
- ($ x(t) $) คือ ตำแหน่งของวัตถุที่เวลา ($ t $)
- ($ x_0 $) คือ ตำแหน่งเริ่มต้น
- ($ v_0 $) คือ ความเร็วเริ่มต้น
- ($ a $) คือ ความเร่ง

2. **โค้ดการจำลอง**:
   ```python
   # กำหนดค่าพื้นฐาน
   x0 = 0  # ตำแหน่งเริ่มต้น
   v0 = 0  # ความเร็วเริ่มต้น
   a = 9.8  # ความเร่ง (อัตราเร่งเนื่องจากแรงดึงดูดของโลก)
   t = np.linspace(0, 10, 100)  # เวลา (จาก 0 ถึง 10 วินาที, จำนวน 100 จุด)

   # การคำนวณตำแหน่งของวัตถุที่เวลา t
   x = x0 + v0 * t + 0.5 * a * t**2

   # แสดงผลการจำลอง
   plt.plot(t, x)
   plt.title('Simulation of Object Motion')
   plt.xlabel('Time [s]')
   plt.ylabel('Position [m]')
   plt.grid(True)
   plt.show()
   ```

#### **ส่วนที่ 3: การจำลองการเคลื่อนที่ในระบบมวล-สปริง**
สมมุติว่าเรามีระบบมวล-สปริง ซึ่งสามารถใช้สมการการเคลื่อนที่ของมวลที่ติดอยู่กับสปริงได้ดังนี้:

$
m \cdot \frac{d^2x}{dt^2} + kx = 0
$
โดยที่:
- ($ m $) คือ มวล
- ($ k $) คือ ค่าความแข็งแรงของสปริง
- ($ x $) คือ การเคลื่อนที่ของมวล

เราจะใช้วิธีการจำลองเชิงตัวเลขในการหาคำตอบจากสมการนี้

1. **โค้ดการจำลอง**:
   ```python
   # ค่าพารามิเตอร์
   m = 1.0  # มวล (กิโลกรัม)
   k = 10.0  # ค่าความแข็งแรงของสปริง (N/m)
   x0 = 1.0  # การยืดออกเริ่มต้น (เมตร)
   v0 = 0.0  # ความเร็วเริ่มต้น (m/s)

   # กำหนดเวลาที่จะทำการจำลอง
   t = np.linspace(0, 10, 500)  # จาก 0 ถึง 10 วินาที
   dt = t[1] - t[0]  # ความแตกต่างของเวลา (dt)

   # สร้างอาร์เรย์เพื่อเก็บค่าการเคลื่อนที่
   x = np.zeros(len(t))
   v = np.zeros(len(t))

   # กำหนดค่าตั้งต้น
   x[0] = x0
   v[0] = v0

   # การจำลองการเคลื่อนที่
   for i in range(1, len(t)):
       # คำนวณความเร่งจากสมการมวล-สปริง
       a = -k * x[i-1] / m
       # ใช้สมการการเคลื่อนที่แบบ Euler
       v[i] = v[i-1] + a * dt
       x[i] = x[i-1] + v[i] * dt

   # แสดงผลการจำลอง
   plt.plot(t, x)
   plt.title('Mass-Spring System Simulation')
   plt.xlabel('Time [s]')
   plt.ylabel('Displacement [m]')
   plt.grid(True)
   plt.show()
   ```

   - คำอธิบาย: กราฟนี้จะแสดงการเคลื่อนที่ของมวลในระบบมวล-สปริง โดยมวลจะสั่นไปรอบ ๆ ตำแหน่งศูนย์กลาง

#### **ส่วนที่ 4: การจำลองการไหลของของเหลว**
จำลองการไหลของของเหลวในท่อที่มีแรงดันและความต้านทาน

1. **สมการการไหล**:
   ใช้สมการของ Poiseuille ซึ่งใช้ในการคำนวณอัตราการไหลของของเหลวในท่อกลม:

$Q = \frac{\pi r^4 \Delta P}{8 \eta L}$

โดยที่:
- ($ Q $) คือ อัตราการไหล (m³/s)
- ($ r $) คือ รัศมีของท่อ (m)
- ($ \Delta P $) คือ ความต่างของความดัน (Pa)
- ($ \eta $) คือ ความหนืดของของเหลว (Pa·s)
- ($ L $) คือ ความยาวของท่อ (m)

2. **โค้ดการจำลอง**:
   ```python
   # กำหนดค่าพารามิเตอร์
   r = 0.05  # รัศมีของท่อ (เมตร)
   delta_P = 500  # ความต่างของความดัน (Pa)
   eta = 0.001  # ความหนืดของของเหลว (Pa·s)
   L = 1.0  # ความยาวของท่อ (เมตร)

   # การคำนวณอัตราการไหล
   Q = (np.pi * r**4 * delta_P) / (8 * eta * L)

   print(f'อัตราการไหล (Q) = {Q:.6f} m³/s')
   ```

   - คำอธิบาย: โค้ดนี้จะคำนวณอัตราการไหลของของเหลวในท่อที่มีรัศมี 0.05 เมตร และค่าความดันต่างกัน 500 Pa
