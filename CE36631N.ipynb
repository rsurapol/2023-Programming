### วิเคราะห์
---

2. input: รับค่า x
3. Process: ทำงานตามเงื่อนไขที่โจทย์กำหนด ดังนี้

Algorithm
---

# บรรยายเอกสาร W06
---


## โจทย์ รับค่าจำนวนเต็ม x เข้ามา 1 จำนวนแล้วหารเอาเศษ(%)ด้วย 3 โดยกำหนดเงื่อนไขดังนี้
* ถ้าเศษเป็นศูนย์ ให้เพิ่ม x เป็น 2 เท่า
* ถ้าเศษเป็นหนึ่งให้เพ่ิม x เป็น 3 เท่า
* ถ้าเศษเป็นสอง ให้บวก x เพิ่มด้วย 5
---


## การวิเคราะห์
---
1. output: แสดงผลค่า x
2. input: รับค่า x
3. process:ทำตามเงื่อนไขโจทย์ดังนี้
* ถ้าเศษเป็นศูนย์ ให้เพิ่ม x เป็น 2 เท่า
* ถ้าเศษเป็นหนึ่งให้เพ่ิม x เป็น 3 เท่า
* ถ้าเศษเป็นสอง ให้บวก x เพิ่มด้วย 5


## การออกแบบ
---
Algorithm
---
1. รับค่าจำนวนเต็ม x
2. ถ้า x%3 เท่ากับ 0
* เป็นจริง x = x * 2
* ไม่เช่นนั้น ถ้า x%3 เท่ากับ 1
* เป็นจริง x = x * 3
* ไม่เช่นนั้น x = x + 5
3. แสดงผล x

Pseudocode
---
```
# มีการจัดรูปแบบเป็นโค้ด
START
      READ x
      IF x%3=0
      THEN x = x * 2
      ELSE IF x%3=1
            THEN x = x * 3
            ELSE x = x + 5
      DISPLAY x
STOP
```



Flochart
---
W06.drawio.png

````python
x = int(input("ป้อนค่าจำนวนเต็ม: "))
if (x%3)==0:
  x = x * 2
elif (x%3)== 1:
  x = x * 3
else:
    x = x + 5

print(x)
````

## EX1-For
---
ให้รับคะแนนของนศ. กลุ่มหนึ่ง แล้วทำการแสดงผล
กำหนดป้อน -1 คือสิ้นสุดการรับคะแนน


## การวิเคราะห์
---
1. output: แสดงผลคะแนนนศ.
2. input: รับคะแนนนศ.
3. process:
>* จำนวนรอบ : ไม่ทราบ
>* เงื่อนไขหยุดวนซ้ำ : ป้อนค่า -1

## การออกแบบ

### Algorithms
---
1. รับคะแนนนศ.
2. วนซ้ำตามเงื่อนไข:หยุดเมื่อ ป้อนคะแนน -1
>* แสดงผลคะแนน
>* รับคะแนนนศ.

### Pseudocode
---


```
# มีการจัดรูปแบบเป็นโค้ด
START
      READ score
      WHILE score != -1:
      THEN
          DISPLAY score
          READ score
STOP
```



### Flowchart
---
ex01-while.drawio.png

````python
#Ex1-While
score = int(input("ป้อนคะแนนนศ.: "))
while score != -1:
  print("คะแนนนศ.: %d"%score)
  score = int(input("ป้อนคะแนนนศ.: "))
````

## EX2-While
---
ให้รับค่าจำนวนจริงค่าหนึ่ง แล้วแสดงผลยกกำลังสอง โดยกำหนดหยุดวนซ้ำ เมื่อป้อนค่าติดลบ

## การวิเคราะห์
---
1. output: แสดงผลยกกำลังสองของตัวเลขที่รับเข้ามา
2. input: รับค่าจำนวนจริงหนึ่งค่า
3. process:
>* จำนวนรอบ: ไม่ทราบ
>* เงื่อนไขหยุด: ป้อนค่าติดลบ

## การออกแบบ

### Algorithms
---
1. รับค่าตัวเลขจำนวนจริงหนึ่งค่า
2. วนซ้ำตามเงื่อนไข: ตัวเลข ไม่เท่ากับ ค่าติดลบ
>* แสดงผลยกกำลังสอง
>* รับค่าตัวเลขจำนวนจริงหนึ่งค่า

### Pseudocode
---


```
# มีการจัดรูปแบบเป็นโค้ด
START
      READ num
      WHILE num > -1:
      THEN
          DISPLAY num*num
          READ num
STOP
```



### Flowchart
---
ex02-while.drawio.png

````python
#EX2-While
num = float(input("ป้อนจำนวนจริง: "))
while num > -1:
  print("ค่ายกกำลัง คือ %.2f"%(num*num))
  num = float(input("ป้อนจำนวนจริง: "))
````
