## โจทย์ จงรับเลขจำนวนเต็ม x เข้ามาแล้วหารเอาเศษ ด้วย 3: กำหนดเงื่อนไขดังนี้
* ถ้าเศษเป็นศูนย์ ให้เพิ่ม x เป็น 2 เท่า
* ถ้าเศษเป็นหนึ่ง ให้เพิ่ม x เป็น 3 เท่า
* ถ้าเศษเป็นสอง ให้บวก x ด้วย 5
---




## การวิเคราะห์
---
1. output: แสดงผลลัพธ์ตามเงื่อนไขโจทย์กำหนด
2. input: รับค่าเลขจำนวนเต็ม x
3. process: ทำตามเงื่อนไขโจทย์กำหนดดังนี้ (x%3)
* ถ้าเศษเป็นศูนย์ ให้เพิ่ม x เป็น 2 เท่า
* ถ้าเศษเป็นหนึ่ง ให้เพิ่ม x เป็น 3 เท่า
* ถ้าเศษเป็นสอง ให้บวก x ด้วย 5


## การออกแบบ
---
Alogorithm:
---
1. รับค่า เลขจำนวนเต็ม x
2. คำนวณ x % 3 แล้วตรวจสอบเงื่อนไข ดังนี้
* ถ้า x % 3 เท่ากับศูนย์ เป็นจริง ให้เพิ่ม x เป็น 2 เท่า
* ไม่เช่นนั้นแล้ว ถ้า x % 3 เท่ากับหนึ่ง เป็นจริง ให้เพิ่ม x เป็น 3 เท่า
* ไม่เช่นนั้นให้ บวก x ด้วย 5
3. แสดงค่า x ตามเงื่อนไข
---

## Pseudocode
---
```
# มีการจัดรูปแบบเป็นโค้ด
START
          READ x
          IF (x%3 = 0)
          THEN x = x * 2
          ELSE IF (x%3 = 1)
                  THEN x =  x * 3
                  ELSE x = x + 5
          DISPLAY x
STOP
```



## Flowchart
---
W06ex1.drawio.png

````python
x = int(input("ป้อนเลขจำนวนเต็ม x"))
if x%3 == 0:
  x = x * 2
elif x%3 == 1:
  x = x * 3
else:
  x = x + 5
print(x)
````

## การวิเคราะห์
---
1. output: แสดงผลรางวัลที่ได้รับ
2. input: รับค่าฉลาก x
3. process: เปรียบเทียบเงื่อนไข x ตามโจทย์กำหนด ดังนี้
* ถ้า 45<=x<=55 เป็นจริง ได้รางวัลที่ 1 (แสดงผลรางวัลที่ 1)
* ถ้า 15<=x<=30 หรือ 75<=x<=90 เป็นจริง ได้รางวัลที่ 2 (แสดงผลรางวัลที่ 2)

## การออกแบบ
---
Algorithm:
---
1. รับค่า x
2. ถ้า 45<=x<=55 เป็นจริง แสดงผลรางวัลที่ 1
 * ไม่เช่นนั้น ถ้า 15<=x<=30 หรือ 75<=x<=90 เป็นจริง แสดงผลรางวัลที่ 2
 * ไม่เช่นนั้น แสดงผลไม่ได้รับรางวัล


## Pseudocode
---

```
# มีการจัดรูปแบบเป็นโค้ด
START
          READ x
          IF (45<=x<=55)
          THEN DISPLAY "First Prize"
          ELSE IF (15<=x<=30) or (75<=x<=90)
                  THEN DISPLAY "Second Prize"
                  ELSE DISPLAY "No Prize"
STOP
```

```
          ELSE IF x<0
                  THEN IF x%2 = 0

                            ELSE DISPLAY x
          ELSE DISPLAY x
STOP
```
